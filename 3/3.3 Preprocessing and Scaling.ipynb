# 3.3 前処理とスケール変換

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython.display import display
from sklearn.model_selection import train_test_split
%matplotlib notebook
%matplotlib inline
```

```python
# 特徴量ごとにスケールを変更してズラした例
# 図の話は3.3.1で
mglearn.plots.plot_scaling()
```

## 3.3.1 さまざまな前処理
上記の左のプロットは2特徴量の2クラス分類合成データセット  
第1の特徴量(x軸)は10から15の間、第2の特徴量(y軸)は1から9の間
右の4つのプロットは、データを標準的なレンジに変換する方法を示している  
sckit-learnにある変換アルゴリズム
1. StandardScaler : 個々の特徴量の平均が0で分散が1になるように変換、全ての特徴量の大きさを揃える、最大値や最小値がその範囲に入ってるわけではない
2. RobustScaler : 個々の特徴量が一定の範囲に入るように変換する、Standarとの違いは中央値と四分位数を用いること、極端に他の値と異なるような値(外れ値)は無視される
3. MinMaxScaler : データが0から1の間に入るように変換
4. Normalizer : 特徴量ベクトルがユークリッド長1になるよう変換

## 3.3.2 データ変換の適用

```python
# cancerデータセットに対して、MinMaxScalerを前処理に用いて、SVMに適用する
# まずはcancerデータセットの確認
# 合計569のデータポイント

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
# MinMaxScalerの場合、fitメソッドは訓練データ中の各特徴量の最小値と最大値を計算
# 教師あり学習とは異なり、X_trainしか与えない
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# transformメソッドを用いて、訓練データをスケール変換する
# 全ての特徴量が0と1の間の値となっている

# データを変換
X_train_scaled = scaler.transform(X_train)
# スケール変換の前後のデータ特性をプリント
print("transformed shape: {}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(X_train_scaled.max(axis=0)))
```

```python
# テストセットのほうもスケール変換を行う
# テストセットの場合は、最小値と最大値が0と1になっていない
# 訓練セットテストセットに対して全く同じ変換をするため
# 訓練データの最小値を引き、訓練データのレンジで割るため

X_test_scaled = scaler.transform(X_test)
# スケール変換の前後のデータ特性をプリント
print("per-feature minimum after scaling:\n {}".format(X_test_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(X_test_scaled.max(axis=0)))
```

## 3.3.3 訓練データとテストデータを同じように変換する

```python
# 最初の図は変換してないもの
# 真ん中の図は同時にスケール変換したもの、グラフ軸が変わっただけ
# 右の図は別々にスケール変換したもの、最小値と最大値が0と1だがデータを勝ってに動かしてしまった

from sklearn.datasets import make_blobs

# 合計データを作成
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
# 訓練セットとデータセットに分割
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

# 訓練セットとテストセットをプロット
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1],
                               c=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^',
                               c=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")

# MinMaxScalerでデータをスケール変換
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# スケール変換されたデータの特性を可視化
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                            c=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1],
                             c=mglearn.cm2(1), label="Test set", s=60)
axes[1].set_title("Scaled Data")

# テストセットを訓練セットとは別にスケール変換
# 最小値と最大値が0, 1になる。ここでは説明のためにわざとやっている
# *実際にはやってはいけない！*
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

# 間違ってスケール変換されたデータを可視化
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label="training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1],
                marker='^', c=mglearn.cm2(1), label="test set", s=60)
axes[2].set_title("Improperly Scaled Data")

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
fig.tight_layout()
```

## 3.3.4 教師あり学習における前処理の効果

```python
# 比較のため、元のデータに対してSVCを適用する

from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    random_state=0)

svm = SVC(C=100)
svm.fit(X_train, y_train)
print("Test set accuracy: {:.2f}".format(svm.score(X_test, y_test)))
```

```python
# MinMaxScalerで前処理する
# スコアが0.63 -> 0.97 に向上

# 0-1スケール変換を使用
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 変換された訓練データで学習
svm.fit(X_train_scaled, y_train)

# スコア
print("Scaled test set accuracy: {:.2f}".format(svm.score(X_test_scaled, y_test)))
```
